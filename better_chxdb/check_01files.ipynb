```python
from sunkit_spex import thermal
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import astropy.units as u
```

```python
thermal.CONTINUUM_GRID=thermal.setup_continuum_parameters('chianti_cont_01_250_unity_v101_t71.sav')
thermal.LINE_GRID=thermal.setup_line_parameters('chianti_lines_01_12_unity_v101_t71.sav')
```

```python
# Maybe energy resolution too high to be trustworthy?
engs=np.arange(0.4,21,0.01)
f10=thermal.thermal_emission(engs << u.keV,10*1e6<<u.K,1e49<< u.cm**(-3))

plt.plot(engs[:-1],f10)
plt.xscale('log')
plt.yscale('log')
plt.xlabel("Energy [keV]")
plt.ylabel("photons s$^{-1}$ cm$^{-2}$ keV$^{-1}$")
plt.show()
```
